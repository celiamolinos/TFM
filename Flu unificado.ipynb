### PoblaciÃ³n Total EspaÃ±a

```python
ine=pd.read_csv("./datos demograficos/spain_ine_1979_2025.csv")
ine['value']=ine.VALUE1*1000000+ine.VALUE2*1000+ine.VALUE3
ine2008=ine.head(43)
ine08=ine2008.sort_values(by="PERIOD", ascending=True)
ine08['AÃ±o'] = pd.to_numeric(ine08['PERIOD'].str[:4]) 
ine08['Trimestre'] = ine08['PERIOD'].str[4:]
```

```python
ine08['mes']=0
for i in range(len(ine08)):
    if ine08.Trimestre[i]=='QI':
        ine08.mes[i]=1
    elif ine08.Trimestre[i]=='QII':
        ine08.mes[i]=4
    elif ine08.Trimestre[i]=='QIII':
        ine08.mes[i]=7
    else:
        ine08.mes[i]=10
```

```python
import datetime
def primer_domingo_tr(anio,mes):
    fecha = datetime.date(anio, mes, 1)  # 1 del mes dado del aÃ±o dado
    dias_hasta_domingo = (6 - fecha.weekday()) % 7  # Calcular dÃ­as hasta el prÃ³ximo domingo
    return fecha + datetime.timedelta(days=dias_hasta_domingo)
```

```python
ine08['Fecha'] = pd.to_datetime(ine08.apply(lambda row: primer_domingo_tr(row['AÃ±o'], row['mes']), axis=1))

ine=ine08.iloc[:,[4,6,10]]
ine.set_index('Fecha', inplace=True)
ine
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(13,5))
sns.lineplot(data=ine,x=ine.index,y=ine.value)
plt.xticks(rotation=90)
plt.show()
```
